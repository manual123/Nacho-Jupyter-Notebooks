# Seaborn relplot

The concern I have with the example gallary code at seaborn.pydata.org is that the code assumes you would be running it in a jupyter notebook environment.  However, if you were to run the example gallery code in a console or in a plain, vanilla .py script, the chart would not get rendered.  As a long time Python user, I know that this code will not render the chart outside of Jupyter notebook environment.  But for a newbie who wants to try out the seaborn code in their .py script will get frustrated or wonder why the chart doesn't show up.  So below is the code taken from the gallery at seaborn.pydata.org, but modified to work in under any context (.py, console, etc).  I also have it so that an actual MATPLOTLIB figure object is created with the .fig attribute because who knows, the user may want to change the figure size maybe?

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the example mpg dataset
mpg = sns.load_dataset("mpg")

# Plot miles per gallon against horsepower with other semantics
fig = sns.relplot(x="horsepower", y="mpg", hue="origin", size="weight",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=6, data=mpg).fig

fig.set_size_inches(8, 7)
plt.title('MPG vs HP by Origin and Weight')
plt.show()
```
