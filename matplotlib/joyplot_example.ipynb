## [JoyPlot](https://github.com/sbebo/joypy) Example

```python
%matplotlib inline
import joypy
import matplotlib.pyplot as plt
import seaborn as sns

# Get sample data
df = sns.load_dataset("tips")

df.head(15)
```

```python
fig, ax = joypy.joyplot(df, column=['total_bill', 'tip'], by='sex', ylim='own', legend=True, grid=True)
fig.set_size_inches(12, 8)
plt.show()
```
