```python
from sympy import symbols, expand, factor, solve, latex, Rational, init_printing
init_printing()  # Activates pretty printing
```

```python
x, y = symbols('x y')
expr = 3 * x **2 + 6 * x + 20
expr
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.array(range(-50,50,1))
y = 3 * x **2 + 6 * x + 20

fig, ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) # set position of x spine to x=0

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))   # set position of y spine to y=0

ax.set_xlim([-10,10])
ax.set_ylim([-1,100])

ax.set_xticks(range(-10,10,1))
ax.set_title(r"$\bf "+latex(expr)+"$", size=20)

ax.grid(True)
ax.plot(x,y)
plt.show()
```
