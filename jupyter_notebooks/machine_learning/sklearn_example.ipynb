```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
plt.rcParams['figure.figsize'] = (8,7)
```

```python
names = [ 'preg' , 'plas' , 'pres' , 'skin' , 'test' , 'mass' , 'pedi' , 'age' , 'category' ]
df = pd.read_csv('pima-indians-diabetes.data.csv', names=names)
```

```python
df.head()
```

preg = # of times pregnant<br>
plas = plasma glucose concentration<br>
pres = diastolic blood pressure (mm Hg)<br>
skin = tricepts skin fold thickness (mm)<br>
test = 2-hour serum insulin (mu U/ml)<br>
mass = body mass index<br>
pedi = diabetes pedigree function<br>
age = age (years)<br>
category = class variable (0 or 1)

# Visualizations

### Histograms

```python
df.hist()
plt.tight_layout()
plt.show()
```

```python
df.plot(kind='density', subplots=True, layout=(3,3), sharex=False)
plt.tight_layout()
plt.show()
```

```python
df.plot.box(subplots=True, layout=(3,3), sharex=False)
plt.tight_layout()
plt.show()
```

```python
sns.heatmap(df.corr(method='pearson'), annot=True)
plt.tight_layout()
plt.yticks(rotation=0)
plt.title("Pearson Correlation Matrix", weight='bold', size=16)
plt.show()
```
