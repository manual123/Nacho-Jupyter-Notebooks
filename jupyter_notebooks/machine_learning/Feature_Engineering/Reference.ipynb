### Min-Max Scaling

$\bar{x}=\frac{x-min(x)}{max(x)-min(x)}$

Min-Max scaling squeezes (or stretches) values to be within a range of values between 0 and 1

```python
df['minmax'] = preproc.minmax_scale(df['n_tokens_content'])
```
