## All Completed Matches To Date

[[back to top](#top)] [[view sample json](#sample)] [[today's matches](#today)] [[tomorrow's matches](#tomorrow)] [[get match events](#events)] [[specific team matches](#team)]

```python
from prettytable import PrettyTable
import simplejson
from urllib import request
json_data = request.urlopen('http://worldcup.sfg.io/matches')
matches = simplejson.load(json_data)

completed_matches = PrettyTable(["Match Number", "Home Team","Home Team Goals", "Away Team", "Away Team Goals", "Winner"])

for match in matches:
    if match['status'] == 'completed':
        match_number = match['match_number']
        home_team = match['home_team']['country']
        away_team = match['away_team']['country']
        home_team_goals = match['home_team']['goals']
        away_team_goals = match['away_team']['goals']
        winner = match['winner']
        completed_matches.add_row([match_number, home_team, home_team_goals, away_team, away_team_goals, winner])

print(completed_matches)
```

## All Today's Matches

[[back to top](#top)] [[view sample json](#sample)] [[all completed](#all)] [[tomorrow's matches](#tomorrow)] [[specific team matches](#team)]

```python
from prettytable import PrettyTable
import datetime
import simplejson
from urllib import request
import dateutil
from pytz import timezone

eastern = timezone('US/Eastern')

json_data = request.urlopen('http://worldcup.sfg.io/matches/today')
today = simplejson.load(json_data)

todays_matches = PrettyTable(["Match Number", "Home Team","Home Team Goals", "Away Team", "Away Team Goals",
                              "Winner", "Status / Start Date+Time"])

for match in today:
    match_number = match['match_number']
    home_team = match['home_team']['country']
    away_team = match['away_team']['country']
    home_team_goals = match['home_team']['goals']
    away_team_goals = match['away_team']['goals']
    if match['winner'] == None:
        winner = 'N/A'
    else:
        winner = match['winner']
        
    if match['status'] == 'future':
        mydate = dateutil.parser.parse(match['datetime'])
        status = mydate.astimezone(eastern).strftime('%Y-%m-%d %H:%M %Z')
    elif match['status'] == 'completed':
        status = 'Completed'
    else:
        status = match['status']
        
    todays_matches.add_row([match_number, home_team, home_team_goals, away_team, away_team_goals, winner, status])

print(todays_matches.get_string(sortby="Status / Start Date+Time"))
```

## Tomorrow's Matches

[[back to top](#top)] [[view sample json](#sample)] [[all completed](#all)] [[today's matches](#today)] [[get match events](#events)] [[specific team matches](#team)]

```python
from prettytable import PrettyTable
import simplejson
from urllib import request
import dateutil
from pytz import timezone

eastern = timezone('US/Eastern')

json_data = request.urlopen('http://worldcup.sfg.io/matches/tomorrow')
tomorrow = simplejson.load(json_data)

tomorrows_matches = PrettyTable(["Home Team", "Away Team", "Start Time"])

for match in tomorrow:
    home_team = match['home_team']['country']
    away_team = match['away_team']['country']
    mydate = dateutil.parser.parse(match['datetime'])
    local_date = mydate.astimezone(eastern).strftime('%Y-%m-%d %H:%M %Z')
    tomorrows_matches.add_row([home_team, away_team, local_date])
    
print(tomorrows_matches.get_string(sortby="Start Time"))
```

## Getting Match Events for a Specific Match Number (game)

[[back to top](#top)] [[view sample json](#sample)] [[all completed](#all)] [[today's matches](#today)] [[tomorrow's matches](#tomorrow)] [[specific team matches](#team)]

```python
from prettytable import PrettyTable
import simplejson
from urllib import request
from pprint import pprint  # Prettier printing of json data
json_data = request.urlopen('http://worldcup.sfg.io/matches')
matches = simplejson.load(json_data)

match_number = input("Enter the match number you want to see the events for: ")

match_events = PrettyTable(["Country","Player", "Time", "Type Of Event"])

for match in matches:
    if match["match_number"] == int(match_number):
        home_team = match["home_team"]["country"]
        away_team = match["away_team"]["country"]
        home_goals = match["home_team"]["goals"]
        away_goals = match["away_team"]["goals"]
        for home_event in match["home_team_events"]:
            player = home_event["player"]
            time = home_event["time"].strip()
            event = home_event["type_of_event"]
            match_events.add_row([home_team, player, time, event])
        
        for away_event in match["away_team_events"]:
            player = away_event["player"]
            time = away_event["time"].strip()
            event = away_event["type_of_event"]
            match_events.add_row([away_team, player, time, event])

print(match_events.get_string(sortby="Time"))
print(home_team, home_goals, ",", away_team, away_goals)
```

## Matches Played by a Specific Team

[[back to top](#top)] [[view sample json](#sample)] [[today's matches](#today)] [[tomorrow's matches](#tomorrow)] [[get match events](#events)] [[all matches completed](#all)]

```python
from prettytable import PrettyTable
import simplejson
from urllib import request
json_data = request.urlopen('http://worldcup.sfg.io/matches')
matches = simplejson.load(json_data)

team = input("Enter team country: ").strip()

team_matches = PrettyTable(["Match Number", "Home Team","Home Team Goals", "Away Team", "Away Team Goals", "Winner"])

for match in matches:
    if match['status'] == 'completed' or match['status'] == 'in progress':
        home_country = match['home_team']['country']
        away_country = match['away_team']['country']
        if home_country == team or away_country == team:
            match_number = match['match_number']
            home_team = match['home_team']['country']
            away_team = match['away_team']['country']
            home_team_goals = match['home_team']['goals']
            away_team_goals = match['away_team']['goals']
            winner = match['winner']
            team_matches.add_row([match_number, home_team, home_team_goals, away_team, away_team_goals, winner])

print(team_matches)
```

### Getting all match events of completed matches

```python
from prettytable import PrettyTable
import simplejson
from urllib import request
import csv
from io import StringIO

json_data = request.urlopen('http://worldcup.sfg.io/matches')
matches = simplejson.load(json_data)

complete_events = PrettyTable(["Match Number", "Team", "Player", "Time", "Type Of Event"])

for match in matches:
    if match['status'] == 'completed':
        home_team = match['home_team']['country']
        away_team = match['away_team']['country']
        match_number = match['match_number']
        for home_event in match['home_team_events']:
            player = home_event['player']
            time = home_event['time']
            event = home_event['type_of_event']
            complete_events.add_row([match_number, home_team, player, time, event])
            
        for away_event in match['away_team_events']:
            player = away_event['player']
            time = away_event['time']
            event = away_event['type_of_event']
            complete_events.add_row([match_number, away_team, player, time, event])

#print(complete_events)
```

```python
column_names = complete_events._get_field_names()
row_data = complete_events._rows

output = StringIO()

writer = csv.writer(output)
writer.writerows([column_names])
writer.writerows(row_data)
events = output.getvalue()
output.close()
```

### Let's see what the match events data set looks like

```python
from datetime import datetime
import pandas as pd

df = pd.read_csv(StringIO(events))
df
```

### Limit the data set to contain just goals

NOTE-there are different types of goals listed: **goal**, **goal-penalty**, and **goal-own** ("own goal').  Unfortunately, the own goal is listed under the team that committed the own goal, not the team that benefitted from the own goal.  So if I were to include own goal in my counts, it would be **INCORRECT!**  So I will create a data set containing "goal" and "goal-penalty".

```python
criteria1 = df['Type Of Event']=='goal'
criteria2 = df['Type Of Event']=='goal-penalty'
```

```python
df_goals = df[criteria1 | criteria2]
```

### Top Goals Scored by Player

Now that I have limited my data to include only goals, I can do goal counts.  Let's see which players have the most goals:

```python
player = df_goals.groupby(['Player'])['Type Of Event'].count()
player.sort(ascending=False)
player
```

### Top Goals Scored by Team/Country

[[back to top](#top)] [[back to summary page](#summary)] [[top player](#player)]

Getting top goal counts by team was a lot trickier.  As already mentioned, I could not use the match events data because of how the own goal was listed.  As a work around, I used the completed matches data instead, since it has the number of goals acquired by each team.  But since there is a separate column for the home team and a separate column for the away team, this complicates things too.  To get around this also, I appended or concatenated a data set containing just the home team+goals with another data set containing just the away team+goals.

```python
# Get the home team columns from the main data set and create a new data set from that
home_team = df_completed[['Home Team', 'Home Team Goals']]

# Get the away team columns from the main data set and create a new data set from that
away_team = df_completed[['Away Team', 'Away Team Goals']]

# In order to concatenate the 2 data sets above, the column names have to be the same,
# so I'll rename the away team's column names
away_team.rename(columns={'Away Team':'Home Team', 'Away Team Goals':'Home Team Goals'}, inplace=True)

# Now I can concatenate the 2 data sets
home_and_away = pd.concat([home_team, away_team])
home_and_away.index = range(0,len(home_and_away))

# I renamed the columns Home Team to just Team and Home Team Goals to just Goals since they are no longer applicable
home_and_away.rename(columns={'Home Team':'Team', 'Home Team Goals':'Goals'}, inplace=True)

# Now I can do a sum of the goals by Team
team = home_and_away.groupby(['Team'])['Goals'].sum()
team.sort(ascending=False)
team
```

```python
print("Total goals scored so far by all teams:", team.sum())
```
