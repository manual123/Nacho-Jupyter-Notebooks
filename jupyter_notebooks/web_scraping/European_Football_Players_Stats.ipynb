```python
import simplejson
from urllib import request
from pprint import pprint  # Prettier printing of json data
json_data = request.urlopen('http://uk.eurosport.yahoo.com/_fantasy_api/resource/salcap.players;game=sogb;split=1;stats=1?crumb=WF3j1BB7uBQ')

players_data = simplejson.load(json_data)
```
