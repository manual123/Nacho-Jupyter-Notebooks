<h2 align='center'>Weibull Q-Q Plotting Example</h2>

<strong>Weibull CDF:</strong> $F(x) = 1 - e^{(\frac{x}\lambda)^k}$<br>
<strong>Inverse Weibull CDF:</strong> $F'(x)=\lambda\left[ln\left(\frac{1}{1-F(x)}\right)\right]^{\frac{1}{k}}$<br><br>
Inverse Weibull CDF was derived by solving for x in the Weibull CDF equation:<br>
<center>$\large{F(x) = 1 - e^{(\frac{x}{\lambda})^k}}$</center><br><br>
<center>$1-F(x)=e^{-(\frac{x}{\lambda})^k}$</center><br><br>
<center>$ln(1-F(x))=-(\frac{x}{\lambda})^k$</center><br><br>
<center>$ln\left(\frac{1}{1-F(x)}\right)=(\frac{x}{\lambda})^k$</center><br><br>
<center>$x=\lambda\left[ln\left(\frac{1}{1-F(x)}\right)\right]^{\frac{1}{k}}$</center>

```python
import numpy as np
from scipy.stats import exponweib
from scipy import stats
from numpy import random
from numpy import log as ln

data = np.array([85,135,150,150,190,200,200,240,240,250])
#data = np.array([384558,483331,508077,515201,615432,666686,726044,755223,807863,848953])
y = ln(data)
rank = np.arange(1,data.size+1)  # ranks = {1, 2, 3, ... 21}
median_rank = (rank - 0.3)/(rank.size + 0.4)
x = ln(-ln(1 - median_rank))

plt.scatter(x,y)
title("Weibull Probability Plot of Data", weight='bold')
grid()
show()

for value in zip(x,y):
    print "( " + str(value[0]) + " , " + str(value[1]) + " )"
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)

line = slope*x+intercept
scatter(x,y)
plot(x,line)
title("Linear Regression - Least Squares Method", weight='bold')
grid()
show()

shape = 1.0/slope
x_intercept = - intercept / shape

print "r value:", r_value
print "slope/shape parameter:", shape
print "intercept:", intercept
scale = exp(-x_intercept/slope)
print "scale parameter:", scale

print '\n'
print "( x, y )"
for value in zip(x,line):
    print "( " + str(value[0]) + " , " + str(value[1]) + " )"
```

```python
x2 = np.arange(0,400)
y2 = 1-exp(-(x2/scale)**shape)

plot(x2,y2)
title("Weibull CDF - Prediction",weight='bold')
grid()
show()

for value in zip(x2,y2):
    print "( " + str(value[0]) + " , " + str(value[1]*100) + " )"
```

Using the equation for the inverse Weibull CDF, we can create a Q-Q plot where the Y-axis is the empirical failure times and the X-axis is the inverse Weibull CDF.<br>
<strong>Inverse Weibull CDF:</strong> $F'(x)=\lambda\left[ln\left(\frac{1}{1-F(x)}\right)\right]^{\frac{1}{k}}$

```python
from numpy import log as ln
import numpy as np
from numpy import random


y = np.array([85,135,150,150,190,200,200,240,240,250])
ranks = np.arange(1,y.size+1)  # ranks = {1, 2, 3, ... 21}
quantiles = (ranks - 0.3)/(ranks.size + 0.4)
x = scale * (ln(1/(1-quantiles)))**(1/shape)


y_theoretical = scale * random.weibull(shape,size=100000)
y_theoretical.sort()
ranks_theoretical = np.arange(1,y_theoretical.size+1)
quantiles_theoretical = (ranks_theoretical - 0.3)/(ranks_theoretical.size + 0.4)
x_theoretical = scale * (ln(1/(1-quantiles_theoretical)))**(1/shape)


plt.scatter(x,y)
plt.plot(x_theoretical, y_theoretical)
title("Data vs Theoretical Weibull Q-Q Plot")
ylabel("Empirical Data")
xlabel("Theoretical Weibull")
show()
```
