<h2 align='center'>Normal Q-Q Plotting Example</h2>

```python
import numpy as np
from scipy.stats import norm
from scipy import stats
from numpy import random

y = np.array([64,64,65,66,66,66,67,67,68,68,68,69,70,70,71,71,72,72,73,74,76])
mu = y.mean()
sigma = y.std()
ranks = np.arange(1,y.size+1)  # ranks = {1, 2, 3, ... 21}
quantiles = (ranks - 0.5)/ranks.size
x = norm.ppf(quantiles)

y_theoretical = random.normal(mu,sigma,size=1000)
y_theoretical.sort()
ranks_theoretical = np.arange(1,y_theoretical.size+1)
quantiles_theoretical = (ranks_theoretical-0.5)/ranks_theoretical.size
x_theoretical = norm.ppf(quantiles_theoretical)

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)

scatter(x,y)
plot(x_theoretical, y_theoretical)
title("Data vs Theoretical Normal Q-Q Plot")
show()
```

```python
line = slope*x+intercept
plot(x,line)
scatter(x,y)
title("Data vs Least Squares Linear Regression Line")
show()
print "r-value:", r_value
print "p-value:", p_value
```

### Now let's see what the distribution would look like using miles to failure.

```python
mtf2009.sort()
data = mtf2009
mu = data.mean()
sigma = data.std()
ranks = np.arange(1,data.size+1)  # ranks = {1, 2, 3, ... 21}
quantiles = (ranks - 0.5)/ranks.size
x = norm.ppf(quantiles)

fig1 = figure(1)
fig1.subplots_adjust(hspace=0.3)
fig1.set_size_inches(12,10)

subplot(2,2,1)
scatter(x,data)
title("Q-Q Plot by MTF",weight="bold")
ylabel("MTF")
xlabel("Quantiles")

subplot(2,2,2)
hist(data,bins=15)
title("Keyless Remote Battery Failures by MTF", weight="bold")

subplot(2,2,4)
boxplot(data, vert=False)
title("Boxplot vs MTF", weight="bold")
show()
```

#### CONCLUSION:

<p>Q-Q plot shows something is wrong at the high miles to fail values.  Histogram also indicates data truncation at the high end of the mile values.  This data truncation makes sense due to the high probability that a vehicle's warranty would typically expire due to mileage before it expires due to days/years.  Looking at the data both by DTF and MTF show that they strongly follow a Normal distribution.  So which measure of failure time should we use to model the failures?  DTF or MTF?  Looking at the data by DTF, it doesn't appear to have experienced data truncation.  Therefore, it would be recommended to use DTF as the failure time, instead of MTF.</p>

### OK, now what?

<p>OK so we've decided to model the data using DTF values instead of MTF.  We also found that we need to remove the early or low DTF values from our data set.  Based on the box plot, most of the low values that it considers "outliers" are values less than 600 DTF.  So we are going to go back to our original data set and grab only those rows where DTF >= 600:</p>

```python
dtf2009 = df[ (df["MODEL_YEAR"]==2009) & (df["DAYS_TO_FAIL_MINZERO"] >= 600)]["DAYS_TO_FAIL_MINZERO"].values
```

#### Now that we have excluded DTF values < 600 days, let's see what the plots look like:

```python
dtf2009.sort()
data = dtf2009
mu = data.mean()
sigma = data.std()
ranks = np.arange(1,data.size+1)  # ranks = {1, 2, 3, ... 21}
quantiles = (ranks - 0.5)/ranks.size
x = norm.ppf(quantiles)

fig1 = figure(1)
fig1.subplots_adjust(hspace=0.3)
fig1.set_size_inches(12,10)

subplot(2,2,1)
scatter(x,data)
title("Q-Q Plot by DTF",weight="bold")
ylabel("DTF")
xlabel("Quantiles")

subplot(2,2,2)
hist(data,bins=20)
title("Keyless Remote Battery Failures by DTF", weight="bold")

subplot(2,2,4)
boxplot(data, vert=False)
title("Boxplot vs DTF", weight="bold")
show()
```

<p>As we can see, the data fits much better onto a straight line in the Q-Q plot.  The histogram looks symetrical and surely looks like a Normal distribution.  Box plot looks nice and symetrical.  Looking good so far!</p>

#### Now, let's see how well the theoretical Normal distribution compares to our warranty data:

```python
from scipy.stats import norm

# Let generate 1000 Normal values with our data's mean (mu) and standard deviation (sigma)
x_theoretical = np.linspace(500,1600,1000)
y_theoretical = norm.pdf(x_theoretical,mu,sigma)

hist(data, bins=20, normed=True)  # Set normed=True
plot(x_theoretical, y_theoretical)
title("2009 Acura TL Keyless Battery vs Normal PDF Plot")
xlabel("DTF")
show()
```

### Looks like a good fit!

#### OK so we have a good model that represents our failure data.  &nbsp Now we can draw inferences from our model.  &nbsp One thing we can do is predict or forecast failures.

#### The histogram and the accompanying Normal PDF above is not useful in making failure forecasts.

<p>This is where the <strong>CDF</strong> (cumulative distribution function) of our Normal data comes in.  Below we'll create the Normal CDF plot:</p>

```python
plot(x_theoretical, norm.cdf(x_theoretical,mu,sigma))
grid()
title("CDF of 09M TL Keyless Battery Failures")
ylabel("Cumulative Percent Failed")
xlabel("DTF")
show()
```

<p>Now that we have a CDF plot, we can then forecast how many failures we can expect after X many days. NOTE-this is <strong>NOT</strong> the same as months or days from start of sales.  That would be asking for a forecast by calendar month.  What we're accomplishing is failure forecast by actual age/failure time.  Please note the difference!  It would take a complex model and significant assumptions to forecast by calendar months since not ALL vehicles are released to the market at the same time.  It has been suggested to use Crow-AMSAA method, but that method is beyond the scope of my knowledge and I don't have a PhD in math or statistics.</p>

### How do we interpret the CDF chart above?

<p>From the CDF chart, we can say that almost 80% of the keyless remote batteries will have failed by 1200 days, almost 60% will have failed by approximately 1,150 days, and so on.  The actual CDF vs DTF values in 30-day intervals are printed below:</p>

```python
days = np.arange(0,1600,step=30)
forecast = norm.cdf(days,mu,sigma)

print "DTF : % Failed"
for item in zip(days,forecast):
    print str(item[0])+" : "+str(item[1]*100)
```
