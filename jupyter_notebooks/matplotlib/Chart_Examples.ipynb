### Histogram

[[Back to Top](#top)]

Example on how to plot data points or values that follow the standard Normal distribution.  
The probability distribution function (pdf) of the standard Normal distribution is:<br>
<center><h2>$f(x,\mu,\sigma)=\frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$</h2></center><br>
where $\mu$ is the *mean* and $\sigma$ is the *standard deviation* of the distribution.

```python
%matplotlib inline
import numpy as np
from numpy import random as nprandom
import matplotlib.pyplot as plt
import seaborn as sns # I prefer seaborn's color scheme to MATPLOTLIB's default
sns.set_style('whitegrid')
sns.set(font="serif") # http://stackoverflow.com/questions/23207516/matplotlib-not-listening-to-font-choices

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

mu, sigma = 0, 1
variates = nprandom.normal(mu,sigma,1000)

# plots the histogram/bar chart
count, bins, ignored = plt.hist(variates, 30, normed=True, color='#3498db', edgecolor='white')

# plots the pdf of the standard Normal distribution per equation above
axes.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
axes.set_title("Histogram of a Standard Normal Variate", weight='bold')
axes.set_xlabel("bins")
axes.set_ylabel("Frequency")
axes.set_facecolor('white')
plt.show()
```

### Box Plot

[[Back to Top](#top)]

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

# Get sample data
tips = sns.load_dataset("tips")

# Make figure and axis with dimension 10x6 inches
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_facecolor('white')

# Create boxplot
sns.boxplot(x="total_bill", y="day", data=tips, ax=ax, orient='h')

# Display the output
plt.show()
```

### Box Plots with Jittered Data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font="serif")

# Get sample data
tips = sns.load_dataset("tips")

# Make figure and axis with dimension 12x8 inches
fig, ax = plt.subplots(figsize=(12, 8))

# Create boxplot and make the facecolor more transparent
sns.boxplot(x="total_bill", y="day", data=tips, ax=ax, orient='h')
for patch in ax.artists:
    r, g, b, a = patch.get_facecolor()
    patch.set_facecolor((r, g, b, 0.6))

# Add strip/jitter plot with semi-transparent data circles
sns.stripplot(x="total_bill", y="day", data=tips,
              size=12, jitter=0.2, edgecolor="gray", alpha=0.6)

# Remove top and right axis spines
sns.despine()

# Override the x and y axis labels
ax.set(xlabel='Total Bill', ylabel='Week Day')
# Add title with specified font size and weight
ax.set_title("Box Plots with Jittered Data Points", fontsize=14, weight='bold')
ax.set_facecolor('white')

# Display the output
plt.show()
```

### Lollipop Plot

[[back to top]](#top)

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(font="serif")
 
# Create a dataframe
df = pd.DataFrame({'group':list(map(chr, range(65, 85))), 'values':np.random.uniform(size=20) })
 
# Reorder it following the values:
ordered_df = df.sort_values(by='values')
my_range=range(1,len(df.index)+1)

fig, axes = plt.subplots(figsize=(10,8))

# The vertival plot is made using the hline function
# I load the seaborn library only to benefit the nice looking feature
plt.hlines(y=my_range, xmin=0, xmax=ordered_df['values'], color='skyblue')
plt.plot(ordered_df['values'], my_range, "o")
 
# Add titles and axis names
plt.yticks(my_range, ordered_df['group'])
plt.title("A vertical lolipop plot", loc='left', weight='bold')
plt.xlabel('Value of the variable')
plt.ylabel('Group')
plt.show()
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font="serif")
 
# Create a dataframe
df = pd.DataFrame({'group':list(map(chr, range(65, 85))), 'values':np.random.uniform(size=20) })
 
# Reorder it following the values:
ordered_df = df.sort_values(by='values')
my_range=range(1,len(df.index)+1)
 
# Create a color if the group is "B"
my_color=np.where(ordered_df ['group']=='B', 'orange', 'skyblue')
my_size=np.where(ordered_df ['group']=='B', 70, 30)

fig, axes = plt.subplots(figsize=(10,8))

# The vertival plot is made using the hline function
# I load the seaborn library only to benefit the nice looking feature
plt.hlines(y=my_range, xmin=0, xmax=ordered_df['values'], color=my_color, alpha=0.4)
plt.scatter(ordered_df['values'], my_range, color=my_color, s=my_size, alpha=1)
 
# Add title and exis names
plt.yticks(my_range, ordered_df['group'])
plt.title("What about the B group?", loc='left', weight='bold')
plt.xlabel('Value of the variable')
plt.ylabel('Group')
plt.show()
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font="serif")
 
# Create a dataframe
value1=np.random.uniform(size=20)
value2=value1+np.random.uniform(size=20)/4
df = pd.DataFrame({'group':list(map(chr, range(65, 85))), 'value1':value1 , 'value2':value2 })
 
# Reorder it following the values of the first value:
ordered_df = df.sort_values(by='value1')
my_range=range(1,len(df.index)+1)

fig, axes = plt.subplots(figsize=(10,8))

# The vertical plot is made using the hline function
# I load the seaborn library only to benefit the nice looking feature
plt.hlines(y=my_range, xmin=ordered_df['value1'], xmax=ordered_df['value2'], color='grey', alpha=0.4)
plt.scatter(ordered_df['value1'], my_range, color='skyblue', alpha=1, label='value1')
plt.scatter(ordered_df['value2'], my_range, color='green', alpha=0.4 , label='value2')
plt.legend()
 
# Add title and axis names
plt.yticks(my_range, ordered_df['group'])
plt.title("Comparison of the value 1 and the value 2", loc='left', weight='bold')
plt.xlabel('Value of the variables')
plt.ylabel('Group')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(font="serif")
 
# Data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x) + np.random.uniform(size=len(x)) - 0.2
 
# Create a color if the group is "B"
my_color=np.where(y>=0, 'orange', 'skyblue')

fig, axes = plt.subplots(figsize=(10, 8))

# The vertival plot is made using the hline function
# I load the seaborn library only to benefit the nice looking feature
plt.vlines(x=x, ymin=0, ymax=y, color=my_color, alpha=0.4)
plt.scatter(x, y, color=my_color, s=1, alpha=1)
 
# Add title and axis names
plt.title("Evolution of the value of ...", loc='left', weight='bold')
plt.xlabel('Value of the variable')
plt.ylabel('Group')
```

### Bubble Plot

[[back top top]](#top)

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(font="serif")
 
# create data
x = np.random.rand(15)
y = x+np.random.rand(15)
z = x+np.random.rand(15)
z=z*z

fig, axes = plt.subplots(figsize=(10, 8))
 
# Change color with c and alpha. I map the color to the X axis value.
plt.scatter(x, y, s=z*2000, c=x, cmap="Blues", alpha=0.4, edgecolors="grey", linewidth=2)
 
# Add titles (main and on axis)
plt.xlabel("the X axis")
plt.ylabel("the Y axis")
plt.title("A colored bubble plot", weight='bold')
 
plt.show()
```

### Seaborn's Violin Plot (a combination of boxplot and kernel density plot)

[[Back to Top](#top)]

```python
%matplotlib inline
import numpy as np
import seaborn as sns

sns.set()

# Create a random dataset across several variables
rs = np.random.RandomState(0)
n, p = 40, 8
d = rs.normal(0, 2, (n, p))
d += np.log(np.arange(1, p + 1)) * -5 + 10

# Use cubehelix to get a custom sequential palette
pal = sns.cubehelix_palette(p, rot=-.5, dark=.3)

# Show each distribution with both violins and points
sns.violinplot(data=d, palette=pal, inner="points")
plt.show()
```

### Stardard Line Plot w/ Classic X and Y axis

[[Back to Top](#top)]

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.array(range(-5,6,1))
y = x**2

fig, ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) # set position of x spine to x=0

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))   # set position of y spine to y=0

#ax.set_xlim([-10,10])
ax.set_ylim([-25,25])

#ax.set_xticks(range(-10,10,1))
ax.set_title("$x^2$", size=24)

ax.grid(True)
ax.plot(x,y)
plt.show()
```

### Scatter Plot

[[Back to Top](#top)]

```python
%matplotlib inline
from numpy import random as nprandom
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # I prefer seaborn's color scheme to MATPLOTLIB's default
sns.set(font="serif") # http://stackoverflow.com/questions/23207516/matplotlib-not-listening-to-font-choices

mu, sigma = 0, 1
variate1 = nprandom.normal(mu,sigma,1000)
variate2 = nprandom.normal(mu,sigma,1000)

data_set = {"data1":variate1,"data2":variate2}
df = pd.DataFrame(data_set)

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

axes.scatter(df["data1"].values, df["data2"].values,)
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("Scatter Plot", weight='bold')
plt.show()
```

### Seaborn's Joint Plot

[[Back to Top](#top)]

What's nice about Seaborn is that it leverages Pandas dataframe data structure:

```python
%matplotlib inline
from numpy import random as nprandom
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # I prefer seaborn's color scheme to MATPLOTLIB's default
sns.set(font="serif") # http://stackoverflow.com/questions/23207516/matplotlib-not-listening-to-font-choices

mu, sigma = 0, 1
variate1 = nprandom.normal(mu,sigma,1000)
variate2 = nprandom.normal(mu,sigma,1000)

data_set = {"data1":variate1,"data2":variate2}
df = pd.DataFrame(data_set)

sns.jointplot("data1","data2",data=df, kind="kde");
plt.show()
```
