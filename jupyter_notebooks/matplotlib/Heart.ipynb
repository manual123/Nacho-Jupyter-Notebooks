### For the nerdy Valentine...

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

t = np.arange(0, 2*np.pi, 0.1)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)

plt.plot(x, y)
plt.fill_between(x, y, y2=5, color='red', data=None)
sns.despine()
plt.axis('off')
plt.show()
```
