<center><h1>Comparing 2 Data Sets Using Python</h1><center>

This jupyter notebook will show how to determine if 2 data sets came from the same distribution using the [Mann-Whitney-Wilcoxon Test](https://en.wikipedia.org/wiki/Mann%E2%80%93Whitney_U_test) or [Kolmogorov-Smirnov 2-Sample Test](https://en.wikipedia.org/wiki/Kolmogorov%E2%80%93Smirnov_test#Two-sample_Kolmogorov.E2.80.93Smirnov_test).  Since the K-S test assumes the distributions of the data sets are from continuous probability distributions, data sets will be drawn from Normal and Weibull distributions.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_excel('/home/pybokeh/Downloads/temp.xlsx', sheetname='Claims')
```

```python
df.boxplot(column='DTF', by='GRADE_SHORT')
plt.show()
```

```python
df.hist(column='DTF', by='GRADE_SHORT')
plt.show()
```

## Mann-Whitney-Wilcoxon Test

```python
import scipy.stats as stats
```

```python
base = df.query("GRADE_SHORT == 'BASE'").MTF
tech = df.query("GRADE_SHORT == 'TECH'").MTF
_stat, _pvalue = stats.mannwhitneyu(base, tech)
```

```python
if _pvalue > 0.05:
    print("Mann-Whitney-Wilcoxon Test result:")
    print("The 2 data sets are not significantly different.  p-value:", _pvalue)
else:
    print("Mann-Whitney-Wilcoxon Test result:")
    print("The 2 data set are significantly different.  p-value:", _pvalue)
```

### Another MWW-test Example

```python
import numpy as np

sample1 = np.random.normal(0,1,size=1000)

plt.hist(x=sample1)
plt.show()
```

```python
scale = 1
shape = 5
sample2 = scale * np.random.weibull(shape, 1000)

plt.hist(x=sample2)
plt.show()
```

```python
_stat, _pvalue = stats.mannwhitneyu(sample1, sample2)
```

```python
if _pvalue > 0.05:
    print("Mann-Whitney-Wilcoxon Test result:")
    print("The 2 data sets are not significantly different.  p-value:", _pvalue)
else:
    print("Mann-Whitney-Wilcoxon Test result:")
    print("The 2 data set are significantly different.  p-value:", _pvalue)
```
