## <center>Forecasting Warranty Defect Rates using Python with a Sigmoid Curve</center>

At work in my warranty department, we have to forecast what is known as cumulative defect rate, which is calculated as cumulative # of claims divided by cumulative # of vehicles sold after some months since the fleet of vehicles have been released to the market.  The x-axis that is representing time, is not failure time, but actually months from start of sales.  Currently, we use Weibull distribution to forecast failure rates based on failure times.  Unfortunately, the management types need to have forecasts based on calendar months.  On top of that, because of warranty limits, the warranty defect rate will wane or level off as vehicles reach their warranty.  Therefore, our <i>warranty</i> defect rate will typically be lower than the <strong>actual</strong> failure rate.  Based on these factors (and others such as seasonality), it is no surprise that forecasting using the Weibull distribution will never match failures based on calendar months.  So thus began my quest to find a better way, a better model.

Based on experience, when the failure of a system is not affected by season, the warranty defect rate almost always follows an S-shaped curved.  I have been looking for a mathematical model that exhibits an S-shaped curve.  After a brief Google search, I stumbled upon the [Sigmoid function](https://en.wikipedia.org/wiki/Sigmoid_function).  

$\huge{f(x)=\frac{1}{1+e^{-x}}}$

#### Using Python, I have charted what the standard Sigmoid function looks like:

```python
import matplotlib.pyplot as plt
from numpy import log as ln
import numpy as np

x = np.array(arange(-6,7,1))
y = 1/(1 + exp(-x))

fig, ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) # set position of x spine to x=0

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))   # set position of y spine to y=0

ax.set_xlim([-6,6])
ax.set_ylim([0,1])

ax.set_xticks(range(-6,7,1))
ax.set_yticks(np.arange(0,1,0.1))

ax.set_ylabel("y")
ax.set_xlabel("x")

ax.grid(True)
ax.plot(x,y)
plt.xticks(rotation=-90)
ax.text(-1.6, 1.1, r"$\frac{1}{1+e^{-x}}$", fontsize=40, color="black")
plt.show()
```

#### However, when dealing with our warranty data, we don't have negative values in the x scale as in the standard Sigmoid function above.  The x axis in our defect rate is based on time.  In my case at work, it is months from start of sales.  So it appears I needed a more generalized version of the Sigmoid function that allows me to shift the function to the positive x-axis direction and to be able to change the shape of the S curve somehow. After some searching I found this [discussion](http://comments.gmane.org/gmane.comp.python.scientific.user/26237) which leads me to find that there are indeed different families/versions of Sigmoid function that use more parameters.  Actually, a Sigmoid is a specific class of [logistic function](https://en.wikipedia.org/wiki/Generalized_logistic_curve).  I have decided that a 4 parameter Sigmoid function is suitable for the warranty data that I work with.

$\huge{f(x)=\frac{U}{(1+e^{r(-x-x0)})}+L}$<br><br>
where U = upper limit, L = lower limit, r = growth rate, and x0 is the x value where growth rate is at its maximum

```python
import matplotlib.pyplot as plt
from numpy import log as ln
import numpy as np

# 4 parameters that I will use to tweak the Sigmoid function
max_growth = 10
growth_rate  = 0.7
upper  = 0.5
lower  = 0

x = np.array(range(-10,50,1))
# The 4-paramter Sigmoid function
y = upper / (1 + np.exp(-growth_rate * (x-max_growth))) + lower

fig, ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) # set position of x spine to x=0

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))   # set position of y spine to y=0

ax.set_xlim([0,20])
ax.set_ylim([0,0.5])

ax.set_xticks(range(0,24,1))
ax.set_yticks(np.arange(0,1,0.1))

ax.set_ylabel("Cumulative Defect %")
ax.set_xlabel("Months From Start of Sales")

ax.grid(True)
ax.plot(x,y)
plt.xticks(rotation=-90)
plt.show()
```

#### The chart above is what a typical cumulative warranty defect rate looks like for a part failure.  The defect rate usually starts off low and steady, then it climbs as the failure mechanism is maturing, then the defect rate level offs as the vehicles are reaching their warranty limit (or it could mean the problem was addressed or it affected only a portion of the fleet of vehicles). So an S-shaped model like the Sigmoid function we just used seems like an excellent choice.  This trend is only valid when the part failure is not influenced by seasonality factors.  For seasonal influences, a different model should be used.  Maybe a double-Sigmoidal function!  Anwyay,that is for another adventure.

#### So the next question is, given some warranty data, how do I estimate the 4 parameters to use in the Sigmoid function?  After some research, I found some [sample code](http://permalink.gmane.org/gmane.comp.python.scientific.user/26242) where we can use Scipy's curve fit function:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def sigmoid(x, x0, k, a, c):
     y = a / (1 + np.exp(-k*(x-x0))) + c
     return y

xdata = np.array([0.0,   1.0,  3.0,  4.3,  7.0,   8.0,   8.5, 10.0,  
12.0, 14.0])
ydata = np.array([0.11, 0.12, 0.14, 0.21, 0.83,  1.45,   1.78,  1.9, 
1.98, 2.02])

popt, pcov = curve_fit(sigmoid, xdata, ydata)
print("Fit:")
print("x0 =", popt[0])
print("k  =", popt[1])
print("a  =", popt[2])
print("c  =", popt[3])
print("Asymptotes are", popt[3], "and", popt[3] + popt[2])

x = np.linspace(-1, 15, 50)
y = sigmoid(x, *popt)

plt.plot(xdata, ydata, 'o', label='data')
plt.plot(x,y, label='fit')
plt.ylim(0, 2.05)
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
