```python
def get_excel(path):
    """Read Excel file containing data we need and converts any spaces in column names to underscores
    
    Parameter
    ---------
    path : str
        - Path to Excel file
        
    Returns
    -------
    dataframe
    """
    
    df = pd.read_excel(path)
    df.columns = [column.replace(' ', '_') for column in df.columns]
    
    
    return pd.read_excel(path)
```

```python
def format_columns(df):
    """Make LAST_NAME and FIRST_NAME all upper case
    
    Parameter
    ---------
    df : dataframe
    
    Returns
    -------
    dataframe
    """
    
    df['LAST_NAME'] = df['LAST_NAME'].str.upper()
    df['FIRST_NAME'] = df['FIRST_NAME'].str.upper()
    
    return df
```

```python
class Export_Plant_Vouchering(object):
    """Create export plant WRP voucher
    """
    
    def __init__(self, plant, path):
        print("Obtaining Excel file...")
        self.plant = plant
        self.df = pd.read_excel(path)
        self.df.columns = [column.replace(' ', '_') for column in self.df.columns]
        print("Finished obtaining Excel file")
        
    def format_data(self):
        self.df['LAST_NAME'] = self.df['LAST_NAME'].str.upper()
        self.df['FIRST_NAME'] = self.df['FIRST_NAME'].str.upper()
        
    def filter_data(self, days_worked, hours_per_day):
        self.df = self.df[(self.df['DAYS_WORKED'] >= int(days_worked)) & (self.df['HOURS_WORKED_PER_DAY'] >= float(hours_per_day))]
        
```
