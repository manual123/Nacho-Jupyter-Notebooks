### Looks like 'dayofweek' attribute is what we need.

### We will use Python dictionary and a function to create a new column containing the day of the week

```python
integer2name = {0:'Monday',1:'Tuesday',2:'Wednesday',3:'Thursday',4:'Friday',5:'Saturday',6:'Sunday'}

def day_of_week(series):
    return integer2name[series['Start Date'].dayofweek]
```

```python
df['Day_Of_Week'] = df.apply(day_of_week,axis=1)
```
