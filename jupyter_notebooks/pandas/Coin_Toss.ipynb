```python
from numpy import random as nprandom
```

```python
heads_or_tails = []
for t in range(1000):
    test = nprandom.random_integers(1,2)
    if test == 1:
        heads_or_tails.append('H')
    else:
        heads_or_tails.append('T')
```

```python
heads_or_tails
```

```python
import pandas as pd

df = pd.DataFrame({'outcome':heads_or_tails}, index=range(1,1001))
```

```python
df.groupby('outcome').size()
```

```python
df.head()
```

```python
def cum_tails(series):
    if series['outcome'] == 'T':
        return 1
    else:
        return 0
```

```python
df["tail_test"] = df.apply(cum_tails, axis=1)
```

```python
df.head()
```

```python
df['cum_tails'] = df['tail_test'].cumsum()
```

```python
df.head()
```

```python
df['pct_ttl'] = (df['cum_tails'] / df.index)*100
```

```python
df.head(10)
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(df.index, df['pct_ttl'])
ax.set_xticks(range(0,1001,50))
ax.set_yticks(np.arange(0,1.1,0.1)*100)
ax.grid(True)
plt.xlabel('# of Coin Tosses')
plt.ylabel('% of Coin Tosses that are Tails')
plt.title('Outcome of Tails after 1000 Coin Tosses', weight='bold')
plt.show()
```
