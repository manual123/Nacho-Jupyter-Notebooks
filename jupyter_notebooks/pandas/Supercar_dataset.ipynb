<center><h1>Analyzing Supercar Data using Python using Pandas</h1></center>

Data sets were provided by [www.sharpsightlabs.com](http://www.sharpsightlabs.com/data-analysis-example-r-supercars-part1/)

```python
import pandas as pd

car_torque = pd.read_csv('http://www.sharpsightlabs.com/wp-content/uploads/2014/11/auto-snout_torque_DATA.txt')
car_0_60_times = pd.read_csv('http://www.sharpsightlabs.com/wp-content/uploads/2014/11/auto-snout_0-60-times_DATA.txt')
# The engine size data set has an extra comma in row 713, so saved a local copy and removed the extra comma
car_engine_size = pd.read_csv('/home/pybokeh/Dropbox/data_sets/supercar/car_engine_size.csv')
# car_engine_size = pd.read_csv('http://www.sharpsightlabs.com/wp-content/uploads/2014/11/auto-snout_engine-size_DATA.txt')
car_horsepower = pd.read_csv('http://www.sharpsightlabs.com/wp-content/uploads/2014/11/auto-snout_horsepower_DATA.txt')
car_top_speed = pd.read_csv('http://www.sharpsightlabs.com/wp-content/uploads/2014/11/auto-snout_top-speed_DATA.txt')
car_power_to_weight = pd.read_csv('http://www.sharpsightlabs.com/wp-content/uploads/2014/11/auto-snout_power-to-weight_DATA.txt')
```

## Let's remove the duplicates

```python
len(car_power_to_weight)
#len(car_torque)
#len(car_0_60_times)
#len(car_engine_size)
#len(car_horsepower)
#len(car_top_speed)
```

```python
car_torque.drop_duplicates(subset='car_full_nm', inplace=True)
car_0_60_times.drop_duplicates(subset='car_full_nm', inplace=True)
car_engine_size.drop_duplicates(subset='car_full_nm', inplace=True)
car_horsepower.drop_duplicates(subset='car_full_nm', inplace=True)
car_top_speed.drop_duplicates(subset='car_full_nm', inplace=True)
car_power_to_weight.drop_duplicates(subset='car_full_nm', inplace=True)
```

```python
pd.set_option("display.max_rows",50)
pd.set_option("display.max_columns",20)
```

```python
car_spec = pd.merge(car_power_to_weight, car_torque, how='left', left_on='car_full_nm', right_on='car_full_nm')
car_spec = pd.merge(car_spec, car_0_60_times, how='left', left_on='car_full_nm', right_on='car_full_nm')
car_spec = pd.merge(car_spec, car_engine_size, how='left', left_on='car_full_nm', right_on='car_full_nm')
car_spec = pd.merge(car_spec, car_horsepower, how='left', left_on='car_full_nm', right_on='car_full_nm')
car_spec = pd.merge(car_spec, car_top_speed, how='left', left_on='car_full_nm', right_on='car_full_nm')
car_spec
```
