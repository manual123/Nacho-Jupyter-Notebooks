## Simple Demos
### Cross-sections (`.xs`)
The cross section method can apply single-equality criteria to any number of levels of an index at the same time (*and* criteria, but not *or*). 

These are very efficient and optimized in pandas and act as vectorized operation on the index

It takes the form: 

`subset = dataframe.xs([val1, val2, ... , valN], level=[col1, col2, ..., colN])`
where `col1` through `colN` are valid index levels

One thing to notice about taking cross sections is that the index levels against which you apply the criteria are dropped (since they are now a single, known value). This is not the case with selections using `.select()` (more on that later).

### So for this, let's select all of the TSS data at the Business

If we were speaking SQL, we'd say:
```
select * 
from raw_data
where location = 'Business'
  and parameter = 'TSS'
```

Here it is in pandas:

````python
business_tss = raw_data.xs(['Business', 'TSS'], level=['location', 'parameter'])
business_tss
````

### Selection (`.select`)
Selections allow us to more arbitrarily apply criteria to the index. This is less effiecient because it has to loop through each row of the index and apply a user-defined seletion function to each row.

To use a selection you have to define a function that will accept a row from the index, inspect it, and decide to either return `True` if you want the data or `False` if you don't.

So let's pull out effluent lead data from either buisness or the airport when the storm was number 3 or higher and influent TSS data from the school. Again, in SQL we'd write:
```
select *
from raw_data
where (
    location in ('aiport', 'business')
and storm >= 3
and station = 'Outflow'
and parameter = 'Lead'
) or (
    location = 'school'
and station = 'Outflow'
and parameter = 'TSS'
)
```

And in pandas...

````python
def criteria(row):
    if (row[0] in ['Airport', 'Business']) & (row[1] >= 3) & (row[2] == 'Outflow') & (row[3] == 'Lead'):
        return True
    elif (row[0] == 'School') & (row[2] == 'Inflow') & (row[3] == 'TSS'):
        return True
    else:
        return False
    
subset = raw_data.select(criteria)
subset 
````

## Advanced Demo
### Premise:
#### We have:
Influent and effluent TSS and lead concentrations measured at 3 different location over the course of multiple storms.
#### We want:
Median influent lead concentration for each location with more than 3 valid datapoints
#### Approach
+ We *cross-section* the rows using the `.xs` method of the dataframe for the parameter and monitoring locations
+ We'll also drop the missing values b/c we don't want them with `.dropna`
+ Aggregate the data by the `location` column to get the number of results
+ Subset the data further by selecting the rows where the data aggregated on `location` has more than three results

````python
# select out just the influent lead data
lead = raw_data.xs(['Lead', 'Inflow'], level=['parameter', 'station'], axis=0)

# remove rows missing data
lead = lead.dropna()

lead
````

## Now we need to count how many valid results we have at each `location`
### Just like in a database, we use a `groupby` statement

````python
all_lead_studies = lead.groupby(level='location').size()

all_lead_studies
````

### Then we can select all the BMPs with more than 3 results

````python
good_lead_studies = all_lead_studies[all_lead_studies >= 3]

good_lead_studies
````

## Here is where the really fancy stuff happens. 
With the `.xs` method, we can really efficiently (computer time) select out single values from an index. But what if we need more?
That's where the `.select` method comes in. `.select` takes in a function that looks at each row of the index, and based on the values at that row, returns either `True` or `False`. Where that function returns `True`, we get data back. Where that function function returns `False`, data are omitted.


````python
def is_bmp_valid(index_row):
    if index_row[0] in good_lead_studies:
        return True
    else:
        return False

lead = lead.select(is_bmp_valid)

## if you know what lambdas are, you can also do:
# lead = lead.select(lambda x: x[0] in good_lead_studies)
lead
````

### Now we can compute the median of the whole dataset *en masse*

````python
lead.median()
````

### Or look at each `location` individually

````python
lead.groupby(level='location').median()
````
