## Read in a CSV instead (`res` = "concentration", units are µg/L)

```python
data = pandas.read_csv('WaterQualityData.csv')
data.head()
```

## Compute the median for each analyte within watersheds

```python
groups = data.groupby(by=['watershed', 'analyte'])
groups.median()
```

## Pivot that data so that analytes are in columns 

```python
groups.median().unstack(level='analyte')
```

## Location at the (first 10) medians by sampling location

```python
data.groupby(by=['sampleloc', 'analyte']).median().unstack(level='analyte').head(10)
```

## Aggregate data on something more complex

```python
def interquartile_range(group):
    '''
    Difference between the 75th and 25th percentiles
    '''
    pctl75 = stats.scoreatpercentile(group, 75)
    pctl25 = stats.scoreatpercentile(group, 25)
    IQR = pctl75 - pctl25
    return IQR


(groups.agg({'res': interquartile_range}) 
       .rename(columns={'res': 'IQR'})
       .unstack(level='watershed')
)
```

## Neat Trick: pandas' `describe` method can be applied and tranposed to a `groupby` object

```python
(groups.apply(lambda g: g.describe().T))
```

## Some seaborn stuff

```python
# log data makes smoother densities
data['logres'] = np.log10(data['res'])
fg = seaborn.FacetGrid(data=data, row='watershed', col='analyte', margin_titles=True)
fg.map(seaborn.kdeplot, 'logres', shade=True)
fg.set_xlabels(r'$\log_{10}$ conc.')
fg.set_ylabels('density')
fg.fig.tight_layout()
```
