```python
import uuid
from rpy2.robjects.packages import importr 
from IPython.core.display import Image
import rpy2.robjects.lib.ggplot2 as ggplot2
import pandas as pd
import pandas.rpy.common as com

# IPython notebooks do not directly support in-line R plots.  This function below is a workaround...
grdevices = importr('grDevices')
def ggplot_notebook(gg, width = 800, height = 600):
    fn = '{uuid}.png'.format(uuid = uuid.uuid4())
    grdevices.png(fn, width = width, height = height)
    gg.plot()
    grdevices.dev_off()
    return Image(filename=fn)

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C':[7,8,9]},index=["one", "two", "three"])

r_dataframe = com.convert_to_r_dataframe(df)
```

```python
p = ggplot2.ggplot(r_dataframe) + \
    ggplot2.aes_string(x='A', y='B') + \
    ggplot2.ggtitle("Plot of a Pandas Dataframe") + \
    ggplot2.geom_point()

ggplot_notebook(p, width=500, height=400)
```

```python
datasets = importr('datasets')

mtcars = datasets.__rdata__.fetch('mtcars')['mtcars']

p = ggplot2.ggplot(mtcars) + \
    ggplot2.ggtitle("Plotting a Built-In R Data Set") + \
    ggplot2.aes_string(x='wt', y='mpg', col='factor(cyl)') + \
    ggplot2.geom_point() + \
    ggplot2.geom_smooth(ggplot2.aes_string(group = 'cyl'),
                         method = 'lm')

ggplot_notebook(p, width=500, height=400)
```
