```python
import pandas as pd
import string
from collections import Counter
from nltk.corpus import stopwords 
pd.set_option('display.max_colwidth', 250)
```

```python
data_set = pd.read_excel('qryDK_A-Rank_Reduced_Columns.xlsx')
```

```python
data_set.head()
```

```python
table = str.maketrans({key: ' ' for key in string.punctuation})
one_big_string = ''
for _, (contdesc) in data_set[['CONTDESC']].iterrows():
    #print(contdesc.values[0].translate(table))
    one_big_string = one_big_string + " " + contdesc.values[0].lower().translate(table)
```

```python
stop_words = set(stopwords.words('english'))
words = one_big_string.split()
word_list = [r for r in words if r not in stop_words]
```

```python
from collections import Counter

wordCount = Counter(word_list)
```

```python
wordCount.most_common(20)
```

```python
def cleanComplaint(row):
    
    table = str.maketrans({key: ' ' for key in string.punctuation.replace('/','')})
    stop_words = set(stopwords.words('english'))
    # Add additional words that are not value-added
    stop_words.add('advise')
    stop_words.add('check')
    stop_words.add('claim')
    stop_words.add('cus')
    stop_words.add('customer')
    stop_words.add('guest')
    stop_words.add('please')
    stop_words.add('sentinel')
    stop_words.add('states')
    
    # Remove words from stop words list
    stop_words.remove('a')
    stop_words.remove('all')
    stop_words.remove('at')
    stop_words.remove('both')
    stop_words.remove('from')
    stop_words.remove('not')
    stop_words.remove('off')
    stop_words.remove('on')
    stop_words.remove('when')
    
    words_list = row.str.lower().str.replace("'",'').str.translate(table).str.split().tolist()
    #print(type(words_list))
    #print(words_list[0])
    words_list_without_stopwords = [word for word in words_list[0] if word not in stop_words]
    sentence = ' '.join(words_list_without_stopwords)
    return sentence
```

```python
data_set['CONTDESC_CLEANED'] = data_set[['CONTDESC']].apply(cleanComplaint, axis='columns')
```
