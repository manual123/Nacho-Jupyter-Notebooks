```python
srs_dict = {'noise_words':['NOIS','NIOS','RATT','RATL','HEARD','HEARS','SQ','VIB','CLICK','TICK','TAP','WHIN','HOWL','WHIRL','LOUD','SOUND'],
            'light_on_words':['LIGHT','LAMP','INDICATOR','LITE',' LIT','LT ON'],
            'inop_words':['INOP','T WORK','T COME','T TURN'],
            'airbag_deployment_words':['DEPLOY','WENT OFF']
           }

engine_dict = {'stalled_words':['STALLED','STALLS','ENGINE DIE','ENG DIE','ENGINE QUIT','CAR DIE','VEHICLE DIE','VEH DIE','LOST POW'],
               'noise_words':['NOIS','NIOS','RATT','RATL','HEARD','HEARS','SQ','VIB','CLICK','TICK','TAP','WHIN','HOWL','WHIRL','LOUD','SOUND'],
              }

def complaint(series):
    for word in srs_dict['airbag_deployment_words']:
        if word in series["TEXT1"] or word in series["TEXT2"]:
            return "DEPLOYMENT"
        
    for word in srs_dict['noise_words']:
        if word in series["TEXT1"] or word in series["TEXT2"]:
            return "NOISE"
        
    for word in srs_dict['inop_words']:
        if word in series["TEXT1"] or word in series["TEXT2"]:
            return "INOP"
        
    for word in srs_dict['light_on_words']:
        if word in series["TEXT1"] or word in series["TEXT2"]:
            return "LIGHT ON"
```

```python
df["COMPLAINT"] = df.apply(complaint,axis=1)
```
