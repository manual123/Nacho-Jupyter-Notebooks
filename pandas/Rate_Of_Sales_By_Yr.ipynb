```python
# https://stackoverflow.com/questions/20965046/pandas-cumulative-sum-and-percentage-on-column
import pyodbc                # Needed to connect to ODBC data sources
import base64                # Needed to de-fuzzy my encrypted password
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from matplotlib.dates import MonthLocator, WeekdayLocator, DateFormatter
from matplotlib.ticker import MultipleLocator
import pandas as pd

pw_file = open(r'D:\webapps\_server\pyodbc\cmq.txt', 'r')
pw = base64.b64decode(pw_file.read()).decode('utf-8')
userid = 'rb10'
pw_file.close()

cnxn_string = 'DSN=CMQ_PROD;UID=' + userid + ';PWD=' + pw

cnxn = pyodbc.connect(cnxn_string)
cursor = cnxn.cursor()

# Get the necessary input from the user
year = input("Enter model year(s) separated by comma: ")
factory = input("Enter factory code(s) separated by comma: ")
model = input("Enter model name(s) separated by comma: ")
dest_code = input("Enter destination code(s), separated by comma: ")

# Make a list for each parameter input
years      = year.split(',')
factories  = factory.split(',')
models     = model.split(',')
dest_codes = dest_code.split(',')

# Now get the number of items in each list so that we can properly use the .join() function below
Y = len(years)       # The number of years the user entered
F = len(factories)   # The number of factories the user entered
M = len(models)      # The number of models the user entered
D = len(dest_codes)  # The number of destination codes the user entered

# In this SQL, there are 5 parameters that we want to get input from the user using IN(%s)
sql = """
SELECT
MDL_YR,
FCTRY_CD,
MDL_NM,
ENG_CYL,
DAYS_TO_SALE,
DTS_BIN,
COUNT(*) AS QTY

FROM (

SELECT
MDL_YR,
FCTRY_CD,
MDL_NM,
ENG_CYL,
DAYS_TO_SALE,

/*** Created bin sizes of 30-day intervals ***/
CASE
WHEN DAYS_TO_SALE BETWEEN 0   AND   30 THEN 1
WHEN DAYS_TO_SALE BETWEEN 31  AND   60 THEN 2
WHEN DAYS_TO_SALE BETWEEN 61  AND   90 THEN 3
WHEN DAYS_TO_SALE BETWEEN 91  AND  120 THEN 4
WHEN DAYS_TO_SALE BETWEEN 121 AND  150 THEN 5
WHEN DAYS_TO_SALE BETWEEN 151 AND  180 THEN 6
WHEN DAYS_TO_SALE BETWEEN 181 AND  210 THEN 7
WHEN DAYS_TO_SALE BETWEEN 211 AND  240 THEN 8
WHEN DAYS_TO_SALE BETWEEN 241 AND  270 THEN 9
WHEN DAYS_TO_SALE BETWEEN 271 AND  300 THEN 10
WHEN DAYS_TO_SALE BETWEEN 301 AND  330 THEN 11
WHEN DAYS_TO_SALE BETWEEN 331 AND  360 THEN 12
WHEN DAYS_TO_SALE > 360 THEN 13
END AS DTS_BIN

FROM (

SELECT
FEATURES.MDL_YR,
FEATURES.FCTRY_CD,
FEATURES.MDL_NM,
FEATURES."ENGINE CYLINDERS" AS ENG_CYL,

/*** Days until sold was defined as retail vehicle sale date minus vehicle production date ***/
CASE
WHEN DAYS(SALE_CAL.CAL_DT) - DAYS(AF_CAL.CAL_DT) < 0 THEN 0
ELSE DAYS(SALE_CAL.CAL_DT) - DAYS(AF_CAL.CAL_DT) END AS DAYS_TO_SALE

FROM CMQ.V_FACT_VHCL_PRDN_SLS PROD

INNER JOIN CMQ.V_DIM_MTO_FEATURE_PIVOT FEATURES
ON PROD.MTO_SK = FEATURES.MTO_SK

INNER JOIN CMQ.V_DIM_DATE AF_CAL
ON PROD.AF_OFF_DT_SK = AF_CAL.CAL_DT_SK

INNER JOIN CMQ.V_DIM_DATE SALE_CAL
ON PROD.RTL_SLS_DT_SK = SALE_CAL.CAL_DT_SK

WHERE
FEATURES.MDL_YR IN(%s)
AND FEATURES.FCTRY_CD IN(%s)
AND FEATURES.MDL_NM IN(%s)
AND FEATURES.DEST_CD IN(%s)
AND PROD.RTL_SLS_DT_SK > 0

) AS TEMP

) AS TEMP

GROUP BY
MDL_YR,
FCTRY_CD,
MDL_NM,
ENG_CYL,
DAYS_TO_SALE,
DTS_BIN
""" % (",".join("?" * Y), ",".join("?" * F), ",".join("?" * M), ",".join("?" * D))
# The join functions above are what makes prepared statements with IN() possible!  R, Y, F, M, and D are the 
# number of values that the user has submitted for those respective ? parameters
# The join() function will generate a string such that we get: ?,?,?,...where the number of ?s are determined
# by how many values that were entered by the user separated by commas

parameter_list = []
for year in years:
    parameter_list.append(year)
    
for factory in factories:
    parameter_list.append(factory)

for model in models:
    parameter_list.append(model)
    
for code in dest_codes:
    parameter_list.append(code)

df = pd.read_sql(sql, cnxn, index_col=None, params=parameter_list)

# Close connections
cursor.close()
cnxn.close()
```

```python
pivoted = pd.pivot_table(df, values="QTY", rows="DTS_BIN", cols="MDL_YR", aggfunc=np.sum).fillna(0)
pivoted
```

```python
pivoted_cum = pd.pivot_table(df, values="QTY", rows="DTS_BIN", cols="MDL_YR", aggfunc=np.sum).fillna(0).cumsum()
pivoted_cum
```

```python
MY2007 = 100*pivoted_cum[2007]/pivoted[2007].sum()
MY2008 = 100*pivoted_cum[2008]/pivoted[2008].sum()
MY2009 = 100*pivoted_cum[2009]/pivoted[2009].sum()
MY2010 = 100*pivoted_cum[2010]/pivoted[2010].sum()
MY2011 = 100*pivoted_cum[2011]/pivoted[2011].sum()
MY2012 = 100*pivoted_cum[2012]/pivoted[2012].sum()
MY2013 = 100*pivoted_cum[2013]/pivoted[2013].sum()
```

```python
cum_percentage = pd.DataFrame({"2007M":MY2007.values,
                               "2008M":MY2008.values,
                               "2009M":MY2009.values,
                               "2010M":MY2010.values,
                               "2011M":MY2011.values,
                               "2012M":MY2012.values,
                               "2013M":MY2013.values},index=pivoted_cum.index)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1.5, 1.5]) # left, bottom, width, height

y_major_ticks = MultipleLocator(5)
ax.yaxis.set_major_locator(y_major_ticks)
ax.yaxis.grid(which='major')

xticks_labels = ["0-30","31-60","61-90","91-120","121-150","151-180","181-210","211-240","241-270",
                 "271-300","301-330","331-360","> 360"]

x_major_ticks = MultipleLocator(1)
ax.xaxis.set_major_locator(x_major_ticks)
ax.xaxis.grid(which='major')

plt.plot(cum_percentage.values)
plt.legend(cum_percentage.columns, loc='best')
plt.title("Honda Fit Rate of Sales vs Days Until Sold", weight='bold', size=14)
plt.xticks(range(len(cum_percentage.index)), xticks_labels, rotation=-90)
plt.xlabel("Days Until Sold BINs", weight='bold')
plt.ylabel("Cumulative Percent of Vehicles SOLD", weight='bold')
plt.show()
```
