```python
from lxml import html
import requests
```

```python
base_url = 'http://espn.go.com'

teams_url = 'http://espn.go.com/nba/teams'

headers = {'user-agent' : 'Mozilla/5.0 pybokeh@protonmail.com'}

teams_page = requests.get(teams_url, headers=headers)
teams_html = html.fromstring(teams_page.content)
```

```python
teams_html
```

```python
teams_html.xpath("//a[contains(@href, '/nba/team/stats/_/name/')]/@href")
```

```python
teams_relative_url = teams_html.xpath("//a[contains(@href, '/nba/team/stats/_/name/')]/@href")
```

```python
len(teams_relative_url)
```

```python
teams_absolute_url = [base_url + url for url in teams_relative_url]
```

```python
teams_absolute_url
```

### Obtain the team code from the URL using the split() function

```python
team_code = [team.split('/')[8] for team in teams_absolute_url]
```

```python
team_code
```

### Obtain the team name from the URL, but replace dashes with space and convert name to title format:

```python
team_name = [team.split('/')[9].replace('-', ' ').title().replace('La ', 'LA ').replace('76E', '76e')
             for team in teams_absolute_url]
```

```python
team_name
```

### Let's create a dictionay that will map the team code to the team name based on our 2 lists we just made

```python
keys = team_code
values = team_name
team_code_to_name = dict(zip(keys, values))
```

```python
team_code_to_name
```

### Let's obtain player names

```python
cle_page = requests.get('http://espn.go.com/nba/team/stats/_/name/cle/cleveland-cavaliers', headers=headers)
cle_html = html.fromstring(cle_page.content)
```

#### The resulting xpath query gives us player names duplicated twice:

```python
cle_html.xpath("//a[@class='AnchorLink' and contains(@href, '/nba/player/_/id/')]/text()")
```

#### So we'll just divide it in half to obtain unique list of player names

```python
player_names_list_dupes = cle_html.xpath("//a[@class='AnchorLink' and contains(@href, '/nba/player/_/id/')]/text()")
```

```python
player_names_list = player_names_list_dupes[:int(len(player_names_list_dupes)/2)]
```

```python
player_names_list
```

```python
len(player_names_list)
```

### Let's obtain player stats

```python
player_stats_idx = list(range(len(player_names_list))) + list(range(len(player_names_list)))
```

```python
player_stats_idx
```

```python
len(player_names_list)
```

```python
player_stats_rows_list = []
for idx in range(len(player_names_list)):
    player_stats_rows_list.append(cle_html.xpath("//section[1]/div[2]/div//tr[@class='Table__TR Table__TR--sm Table__even' and @data-idx='" + str(idx) +"']/td/span/text()"))
```

```python
len(player_stats_rows_list)
```

```python
player_stats_rows_list
```

```python
shooting_stats_rows_list = []
for idx in range(len(player_names_list)):
    shooting_stats_rows_list.append(cle_html.xpath("//section[2]/div[2]/div//tr[@class='Table__TR Table__TR--sm Table__even' and @data-idx='" + str(idx) +"']/td/span/text()"))
```

```python
len(shooting_stats_rows_list)
```
