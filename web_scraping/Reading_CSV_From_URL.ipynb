## Reading a csv file using urllib

```python
import csv
from urllib import request
import codecs

input_file = request.urlopen('https://dl.dropboxusercontent.com/u/1929216/airport-codes.csv')

for row in csv.reader(codecs.iterdecode(input_file,'utf-8'), delimiter=','):
    print(row[1])
```
