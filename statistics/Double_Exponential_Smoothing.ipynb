<center>$\large{forecast_t = level_{t-1} + trend_{t-1}}$<br><br>
$\large{level_t = level_{t-1} + trend_{t-1} + \alpha*(forecast_t - (level_{t-1} + trend_{t-1}))}$<br>
or more generically:<br>
$\large{level_t = level_{t-1} + trend_{t-1} + \alpha*(forecast \space error_t)}$<br><br>
$\large{trend_t = trend_{t-1} + \gamma*\alpha*(forecast_t - (level_{t-1} - trend_{t-1}))}$<br>
or more generically:<br>
$\large{trend_t = trend_{t-1} + \gamma*\alpha*(forecast \space error_t)}$<br><br>
$\large{squared \space error = forecast \space error^2}$<br><br>
$\large{standard \space error = \sqrt{\frac{\sum squared \space error}{n - 2}}}$</center><br><br>
Ideally, we want alpha and gamma values to be such that the standard error is at its minimum.

```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
import numpy as np
from math import sqrt

df = pd.read_csv('/home/pybokeh/Dropbox/data_sets/demand.txt', names=['Demand'])

df.index = range(1,len(df)+1)  # by default, the index starts with 0, but want to change it to start with one

# I converted my values to numpy arrays for performance reason
x = np.array(df.index)
y = np.array([value[0] for value in df.values])
```

```python
plt.plot(x,y)
plt.title("Actual Demand")
plt.grid(True)
plt.show()
```

```python
initial_data_size = input("How many data points you want to use to calculate the initial level and trend? ")

t0, l0, r_value, p_value, std_err = stats.linregress(x[:int(initial_data_size)],y[:int(initial_data_size)])

print("initial level:", l0)
print("initial trend:", t0)
```

```python
y0 = t0 * x + l0

plt.plot(x,y,x,y0)
plt.title("Actual Demand with Least Squares Regression of 1st " + initial_data_size + " data points")
plt.grid(True)
plt.show()
```

#### Below is my attempt of translating the equations above into Python data structures, looping through all possible valid values of alpha and gamma whereby in the end, it returns the alpha and gamma combination with the smallest standard error.  I know this is not a very efficient algorithm and use of Python data structures.  This code will run very slow if I increase the alpha/gamma value calculation out to 3 decimal places.  For now, I am just obtaining 2 decimal places of accuracy.  I will try to refactor the code later when I have the time.

```python
SE_list = []
for alpha in np.arange(0.01,1.01,0.01):
    for gamma in np.arange(0.01,1.01,0.01):
        level = []
        trend = []
        level.append(l0)
        trend.append(t0)

        forecast = []
        forecast_error = []
        squared_error = []

        for i, demand in enumerate(y):
            forecast.insert(i,level[-1] + trend[-1])
            forecast_error.insert(i,demand - forecast[i])
            level.insert(i+1,level[-1]+trend[-1]+alpha*forecast_error[-1])
            trend.insert(i+1,trend[-1]+alpha*gamma*forecast_error[-1])
            squared_error.insert(i,forecast_error[i]**2)
        
        SE = sqrt(np.array(squared_error).sum() / ( len(y) - 2 ) )
        SE_list.append([SE, alpha, gamma])
        
SE_alpha_gamma = min(SE_list)  # grab the smallest SE and its corresponding alpha and gamma values
print("alpha and gamma values where standard error is at a minimum:", SE_alpha_gamma)
```

#### Using the optimal alpha and gamma values that were calculated from above, let's provide a forecast going out to 12 data points of time.

```python
level = []
trend = []
level.append(l0)
trend.append(t0)

forecast = []
forecast_error = []
squared_error = []

alpha = SE_alpha_gamma[1]  # get the alpha value that returns minimum standard error
gamma = SE_alpha_gamma[2]  # get the gamma value that returns minimum standard error

for i, demand in enumerate(y):
    forecast.insert(i,level[-1] + trend[-1])
    forecast_error.insert(i,demand - forecast[i])
    level.insert(i+1,level[-1]+trend[-1]+alpha*forecast_error[-1])
    trend.insert(i+1,trend[-1]+alpha*gamma*forecast_error[-1])
    squared_error.insert(i,forecast_error[i]**2)

forecast_length = input("Enter how many months you want to forecast into the future: ")
forecast_periods = range(1,int(forecast_length)+1)

forecast_data = []
for period in forecast_periods:
    print(period, level[-1] + period * trend[-1])
    forecast_data.append(level[-1] + period * trend[-1])
```

#### Now, let's plot the 12 forecast demand values along with our actual demand:

```python
x1 = range(len(y)+1, len(y) + len(forecast_data) + 1, 1)
plt.plot(x,y,x1,forecast_data)
plt.title("Actual Demand with Forecast Trend")
plt.grid(True)
plt.show()
```
