### How-to DOS batch file to activate a Python environment, execute Python script, and deactivate environment

**Step 1:** Create a DOS batch (.bat) file per this StackOverflow [question](https://stackoverflow.com/questions/69068/long-commands-split-over-multiple-lines-in-windows-vista-batch-bat-file), you use ```space``` + caret symbol to run a command that spans multiple lines.

Contents of "test.bat":

@echo off<br>
d:<br>
D:\Miniconda3_32bit\condabin\activate.bat pydata && ^<br>
D:\Miniconda3_32bit\envs\pydata\python.exe D:\jupyter\BRG_BRAIN\workshops\Intro_To_Python\etl.py && ^<br>
D:\Miniconda3_32bit\condabin\conda.bat deactivate && ^<br>
D:\Miniconda3\envs\notebook\Scripts\deactivate.bat && ^<br>
exit /B

The above does the following:
- Change to D: drive (by default, my DOS terminal is first launched in Y: drive, may be different for you)
- Activate my virtual environment
- Change to the directory that has my Python script
- Execute the Python script
- Deactivate the virtual environment
- exit
- Try to Close Command Prompt window

**Step 2:** Add a new "Basic Task" in Windows Task Scheduler and reference the .bat file in Windows Scheduler in the ```Actions``` tab by going through the "Create Basic Task..." dialog wizard and then you're done!
